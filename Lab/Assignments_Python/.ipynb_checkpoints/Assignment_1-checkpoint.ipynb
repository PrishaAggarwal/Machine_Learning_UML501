(f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find <br>
i. Sum of diagonal elements <br>
ii. Eigen values of matrix <br>
iii. Eigen vectors of matrix <br>
iv. Inverse of matrix <br>
v. Determinant of matrix <br>

```python
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
diag_sum = np.trace(n_array) #1
print(diag_sum)

e_val, e_vec = np.linalg.eig(n_array)
print(e_val,e_vec) #2,3

inverse = np.linalg.inv(n_array)
print(inverse) #4

determinant = np.linalg.det(n_array)
print(determinant) #5
```

(g) Multiply the following matrices and also find covariance between matrices using NumPy: <br>
i. <br>
p = [[1, 2], [2, 3]] <br>
q = [[4, 5], [6, 7]] <br>
ii. <br>
p = [[1, 2], [2, 3], [4, 5]] <br>
q = [[4, 5, 1], [6, 7, 2]] <br>

```python
p = np.array([[1,2],[2,3]])
q = np.array([[4,5],[6,7]])
product = np.dot(p,q)
cov = np.cov(p.T,q.T)
print(product,cov)  #1

p = np.array([[1,2],[2,3],[4,5]])
q = np.array([[4,5,1],[6,7,2]])
product2 = np.dot(p,q)
cov2 = np.cov(p.flatten(), q.flatten())
print(product2,cov2) #2
```

(h) For the matrices: x = np.array([[2, 3, 4], [3, 2, 9]]); y = np.array([[1, 5, 0], [5, 10, 3]]), 
find inner, outer and cartesian product?

```python
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

inner = np.inner(x, y)

outer = np.outer(x.flatten(), y.flatten())

cartesian = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)

print(inner, outer, cartesian)
```

Q2: Based on NumPy Mathematics and Statistics <br>
(a) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]]) <br>
i. Find element-wise absolute value <br>
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each 
row. <br>
iii. Mean, Median and Standard Deviation of flattened array, of each column, and 
each row. <br>

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

a_arr = np.abs(array)
print("absolute values : ",a_arr) #1

flat = array.flatten()
p_flat = np.percentile(flat,[25,50,75])
p_col = np.percentile(array,[25,50,75],axis=0)
p_row = np.percentile(array,[25,50,75],axis=1)
print("Percentiles Flattened:", p_flat)
print("Percentiles Col-wise: ", p_col)
print("Percentiles Row-wise : ", p_row) #2


mean_flat = np.mean(flat)
median_flat = np.median(flat)
std_flat = np.std(flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("Flattened Mean:", mean_flat, "Median:", median_flat, "Std:", std_flat)
print("Col wise  Mean:", mean_col, "\nMedian:", median_col, "\nStd:", std_col)
print("Row wise  Mean:", mean_row, "\nMedian:", median_row, "\nStd:", std_row) #3
```

(b) For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]). Find floor, ceiling and truncated value, rounded values

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
f = np.floor(a)
c = np.ceil(a)
t = np.trunc(a)
r = np.round(a)
print("Floor:", f)
print("Ceiling:", c)
print("Truncated:", t)
print("Rounded:", r)
```

Q3: Based on Searching and Sorting <br>
(a) For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find <br>
i. Sorted array <br>
ii. Indices of sorted array <br>
iii. 4 smallest elements <br>
iv. 5 largest elements <br>

```python
array = np.array([10,52,62,16,16,54,453])

s_array = np.sort(array)
print("Sorted array:", s_array) #1

s_indices = np.argsort(array)
print("Indices of sorted array:", s_indices) #2

s_4 = np.sort(array)[:4]
print("4 smallest elements:", s_4) #3

l_5 = np.sort(array)[-5:][::-1]
print("5 largest elements:", l_5) #4
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find <br>
i. Integer elements only <br>
ii. Float elements only <br>

```python
array = np.array([1.0,1.2,2.2,2.0,3.0,2.0])

int_el = array[array == array.astype(int)]
print("Integer elements only:", int_el) #1

f_el = array[array != array.astype(int)]
print("Float elements only:", f_el) #2
```

(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images) <br>
(b) Load the saved file into jupyter notebook?

```python
!pip install opencv-python
import cv2
```

```python
def img_to_array(path, output='image_arr.txt'):
    img = cv2.imread(path)

    if img is None:
        raise ValueError("Image not found or path is incorrect.")

    if len(img.shape) == 2 or img.shape[2] == 1:
        print("Grayscale image detected.")
        gray_img = img if len(img.shape) == 2 else cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        np.savetxt(output, gray_img, fmt='%d')
    else:
        print("RGB image detected.")
        # Save each channel separately
        for i, color in enumerate(['B', 'G', 'R']):
            channel = img[:, :, i]
            np.savetxt(f'{color}_{output}', channel, fmt='%d')

    print("Image saved as text file(s).") #1

gray = np.loadtxt('image_array.txt', dtype=np.uint8)
print("Grayscale array loaded:\n", gray) #2

b = np.loadtxt('B_image_array.txt', dtype=np.uint8)
g = np.loadtxt('G_image_array.txt', dtype=np.uint8)
r = np.loadtxt('R_image_array.txt', dtype=np.uint8)

# Reconstruct image
reconstructed_img = cv2.merge([b, g, r])
print("RGB array reconstructed.")
```
